```python
df = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])
df
```

```python
n=4 
# view the front n lines
dfh = df.head(n)
print(dfh)

# view the last n lines
dft = df.tail(n) 
print(dfh)

# 各类统计信息
dfs = df.describe() 
print(dfh)

# assign rows and columns to with iloc[ , ] 
top_left_corner_df = df.iloc[:5, :5]
print(top_left_corner_df)

# transpose rows and cols
dfT = df.T 
print(dfT)
```

```python
# Some attributes of df

# list row and col indexes
l = df.axes
print(l)

# from above??????
(r, c) = df.axes 
print((r,c))

# Series column data types
s = df.dtypes
print(s)

# True for empty DataFrame
b = df.empty
print(b)

# number of axes (2)
i = df.ndim 
print(i)

 # (row-count, column-count)
t = df.shape
print(t)

# from above
(r, c) = df.shape 
print((r,c))

# row-count * column-count
i = df.size 
print(i)

# get a numpy array for df
a = df.values 
print(a)
```

```python
# label, list

df = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])
idx = df.columns  # get col index
label = df.columns[0:3]  # 1st col label
lst = df.columns.tolist()  # get as a list
print(idx)
print(label)
print(lst)
```
